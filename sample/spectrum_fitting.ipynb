```python
data = np.load('data/sample_spectrum.npy')
wavelength = data[0]
spectrum = data[1]
data2 = np.load('data/sample_spectrum2.npy')
wavelength2 = data2[0]
spectrum2 = data2[1]

plt.plot(wavelength, spectrum)
plt.xlabel('wavelength (nm)')
```

# ガウス関数でフィッティング
## 1つピーク

```python
result = fit.single_peak_fit(wavelength, spectrum, peak_x=632.05)
pprint.pprint(result)
```

## 2つピーク

```python
result = fit.double_peak_fit(wavelength, spectrum, peak_x=632.93, other_peak=632.97)
pprint.pprint(result)
```

## 3つピーク

```python
result = fit.triple_peak_fit(wavelength2, spectrum2, peak_x=609.04, other_peak=[609.09, 608.98])
pprint.pprint(result)
```
