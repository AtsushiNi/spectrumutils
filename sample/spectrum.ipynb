# テストデータ

```python
data = np.load('data/spectrum_array.npy')
```

```python
wavelengths = data[0]
spectra = data[1]

plt.figure(figsize=(16,8))
for i in np.arange(wavelengths.shape[0]):
    plt.plot(wavelengths[i], spectra[i])
```

# スペクトルの重なりを除去して一本のスペクトルデータにする

```python
wavelength, spectrum = sp.unify2(wavelengths, spectra)
```

```python
plt.figure(figsize=(16,8))
plt.plot(wavelength, spectrum)
```
